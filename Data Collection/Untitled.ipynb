```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# Scrapping my Accomplishments from Coursera.

```python
url='https://www.coursera.org/account/accomplishments/professional-cert/G5T59BTQSMUU'
webpage = requests.get(url).text
```

```python
soup = BeautifulSoup(webpage)
```

```python
print(soup.prettify())
```

```python
len(soup.find_all('div',class_='course-record-tile-details'))
```

```python
for i in soup.find_all('div',class_='course-record-tile-details'):
    name = i.find('h3')
    print(name.text)
```
